```python
from sklearn import tree
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df = df.dropna(subset=['Survived','Age'])
df = pd.get_dummies(df, columns=["Sex", "Pclass"], drop_first=True)
df = pd.get_dummies(df, columns=["Embarked"], drop_first = True)
df.sample(5)
```

```python
L=["Embarked_Q", "Embarked_S", "Sex_male", "Pclass_2.0", "Pclass_3.0", "Age"]
x = df[["Embarked_Q", "Embarked_S", "Sex_male", "Pclass_2.0", "Pclass_3.0", "Age"]]
y = df.Survived
clf = tree.DecisionTreeClassifier(max_depth=4)
clf = clf.fit(x,y)
```

```python
data = tree.export_graphviz(clf,out_file="titanictree.dot")
```

We definitely see some overfitting in the leafs, especially near the end on the left.
