```python
import xgboost
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from keras.datasets import fashion_mnist
from sklearn.preprocessing import StandardScaler
import numpy as np
```

Testing XGBoost with a grid search on the Fashion MNIST dataset

The MSE is a bad choice for this dataset because this dataset is a bunch of images. The MSE would only push us toward the average pixel intensity.
A better objective would be softmax, which would put the data into multiple classes.

```python
(X_train,y_train),(X_test,y_test) = fashion_mnist.load_data()
ones2get = np.random.randint(0,len(X_train),6000)
X_train = X_train.reshape(-1,28**2)[ones2get]
y_train = y_train[ones2get]
X_test = X_test.reshape(-1,28**2)
y_test = y_test
```

```python
pipe = Pipeline([('scaler',StandardScaler()),('xgbc',xgboost.XGBClassifier(objective='softmax'))])
param_grid = [{'xgbc__eta':[0.3,0.6],
               'xgbc__alpha':[0.5,1],
               'xgbc__gamma':[5,0.8],
              'xgbc__lambda':[3,0.6]}]
gs = GridSearchCV(pipe, param_grid, cv=3, verbose=1,n_jobs=-1).fit(X_train,y_train)
```

```python
print(gs.best_params_,gs.best_score_)
```

```python
pipe = Pipeline([('scaler',StandardScaler()),('xgbc',xgboost.XGBRegressor(objective='softmax'))])
param_grid = [{'xgbc__eta':[0.3,0.7],
               'xgbc__alpha':[0.5,1],
               'xgbc__gamma':[0.5,0.8],
              'xgbc__lambda':[0.3,0.6]}]
gs = GridSearchCV(pipe, param_grid, cv=3, verbose=1,n_jobs=-1).fit(X_train,y_train)
print(gs.best_params_,gs.best_score_)
```
