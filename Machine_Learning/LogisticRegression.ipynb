```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import Logit
from scipy import optimize
import statsmodels.api as sm
```

```python
df = pd.read_csv("titanic.csv")
df = df.dropna(subset=['Survived'])
df.head()
```

```python
df = pd.get_dummies(df, columns=["Embarked"], drop_first = True)
df.head()
```

```python
X= df[["Embarked_Q", "Embarked_S"]]
Y = df["Survived"]
X = sm.add_constant(X)
logit_model = Logit(Y,X).fit()
print(logit_model.summary2())
```

All city vars are statistically significant

```python
df = pd.get_dummies(df, columns=["Sex", "Pclass"], drop_first=True)
df = df.dropna(subset=["Embarked_Q", "Embarked_S", "Sex_male", "Pclass_2.0", "Pclass_3.0", "Age", "Fare"])
df.head()
```

```python
X = df[["Embarked_Q", "Embarked_S", "Sex_male", "Pclass_2.0", "Pclass_3.0", "Age", "Fare"]]
Y = df["Survived"]
X = sm.add_constant(X)
logit_model = Logit(Y,X).fit()
print(logit_model.summary2())
```

All features except Fare are statistically significant

They are significant

```python
class logistic_regression:
    def __init__(self, X, Y):
        """
        Parameters:
            X (N,D) ndarray
            Y (N,1) ndarray: Each element is 0 or 1
        """
        self.X = X
        self.Y = Y
        self.w=None
        
    def fit(self):
        """Returns a fit"""
        s = lambda x: np.exp(x) / (1+np.exp(x))
        def neg_likelihood(w):
            wTx = np.array([w@x for x in X])
            return -np.sum(self.Y*np.log(s(wTx))+(1-self.Y)*np.log(1-s(wTx)))
        initial_guess = sm.OLS(Y,X).fit().params
        w = optimize.minimize(neg_likelihood, initial_guess,tol=.0001).x
        self.w=w
        return w
        
    def predict_proba(self,x):
        """Predicts the probability of y given x"""
        if self.w:
            return 1/(1+np.exp(-np.dot(self.w,x)))
        else:
            raise NotImplemented('You have not fit the data yet')
        
        
    def predict(self,x):
        """Returns a true or false prediction value based on the value of our prediction"""
        prob=self.predict_proba(x)
        if prob>.5:
            return 1
        else:
            return 0
```

```python
df["Constant"] = 1
X = df[["Constant", "Embarked_Q", "Embarked_S", "Sex_male", "Pclass_2.0", "Pclass_3.0", "Age", "Fare"]].values
Y = df["Survived"].values
regressor = logistic_regression(X, Y)
coeff = regressor.fit()
print(coeff)
```
