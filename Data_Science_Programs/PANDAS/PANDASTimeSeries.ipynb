The file `website_traffic.csv` contains records for different visits to a fictitious website.

```python
traffic = pd.read_csv("website_traffic.csv")
#read it in and change enter and leave to datetime objects
traffic.ENTER = pd.to_datetime(traffic.ENTER)
traffic.LEAVE = pd.to_datetime(traffic.LEAVE)
#make a third column being duration of time spent on site
g=traffic.LEAVE-traffic.ENTER
traffic['DURATION']=g
traffic.index = traffic.ENTER
traffic
```

```python
#resample according to minutes and calculate the average visit duration by minute
minutes = traffic.DURATION.resample("T")
mins = minutes.sum()/minutes.agg(len)
#resample according to hours and calculate the avg visit duration by hour
hours = traffic.DURATION.resample("H")
hrs = hours.sum()/hours.agg(len)
#plot the results
plt.figure()
plt.ylabel("Duration")
plt.title("Downsampled Average Visit Duration")
mins.plot(color='g',label="Aggregated by Minutes")
hrs.plot(color='r',label="Aggregated by Hours")
plt.legend(loc='best')
plt.xlabel("Time of Access")
plt.show()
```
